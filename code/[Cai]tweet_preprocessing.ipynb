```python
import os
filename = os.listdir('/content/drive/MyDrive/twitter-celebrity-tweets-data/')
```

```python
import re
import nltk
nltk.download('stopwords')
from nltk.stem import PorterStemmer
from nltk.corpus import stopwords
import pandas as pd
from tqdm import tqdm

porter_stemmer = PorterStemmer()
stop_words = set(stopwords.words('english'))
```

```python
def preprocessor(text):
    text = ' '.join([i for i in text])
    return text
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer(max_features=1000,stop_words='english',lowercase=False,use_idf=True,norm='l2',smooth_idf=True)
```

```python
import pandas as pd
for abc in filename:   
    # if not abc.startswith('.'):
    #     word = re.findall(r"\w+", open('./data'+'//'+abc,'r').read().lower())  
    data = pd.read_csv('/content/drive/MyDrive/tmp/'+abc)
    # display(data)
    data_tweet = data.copy()
    # # tokens = []
    for d in range(len(data)):
      post = data.loc[d,'tweet']
      tokens = []
      post = re.sub(r'\|\|\|', r"    ", post)    # 去除'|||'
      post = re.sub(r'http\S+', r' ', post)    # 去除網址
      post = re.sub(r'\d','', post)  # 去除數字
      # post = re.sub(r'[.,"\-?:!;@]', '', post)
      post = re.sub(r'^b\'', '', post) 
      post = re.sub(r'x[A-Za-z]', '', post)
      post = re.sub(r'\b[A-Za-z]\b', '', post)

      # 取出所有格 -> 放入列表 -> 去除所有格
      x = re.findall("[A-Za-z]+\'+\w+", post)   # x is a list
      for i in x:
          tokens.append(i)
      post = re.sub(r"[A-Za-z]+\'+\w+", r"   ", post)  # 去除所有格

      result = re.findall( r"\w+" , post.lower())  # tokenizaion

      # for i in result:
      #     tokens.append(i)
      word = []
      for data1 in result:
          if data1 not in stop_words:
              word.append(data1)
      for i in word:
        i = re.sub(r'[.,"\-?:!;@_]', '', i)
        tokens.append(porter_stemmer.stem(i))
      
      # print(tokens)
    #   # 整理回 dataframe
      data_tweet.tweet[d] = tokens
    for i in tqdm(range(data_tweet.shape[0])):
      data_tweet.loc[i,'text'] = preprocessor(data_tweet['tweet'][i])
    
    x = tfidf.fit_transform(data_tweet.text)
    vec_text = x.toarray()
    vec_df = pd.DataFrame(vec_text)
    vec_df.to_csv('/content/drive/MyDrive/tweet_processed/'+abc)
```
